```python
import pandas as pd
import great_expectations as gx
```

## Memuat Data

```python
# Load Data
file_name = 'P2M3_arief_bagus_nugraha_data_raw.csv'
df = pd.read_csv(file_name)

# Tampilkan 5 baris pertama untuk memastikan data terbaca
df.head()
```

```python
# Info Table
df.info()
```

## Persiapan untuk validasi

```python
# Membuat Transaction_ID (Gabungan Product_Id + Index) agar Unik
df['Transaction_ID'] = df['Product_Id'] + '_' + df.index.astype(str)

# Mengmbil kategori utama saja (Split string dan ambil elemen pertama)
df['Main_Category'] = df['Category'].str.split('|').str[0]

# Memastikan Discounted_Price adalah float
df['Discounted_Price'] = df['Discounted_Price'].astype(float)

print("Preprocessing selesai. Kolom baru terbentuk.")
df[['Transaction_ID', 'Main_Category', 'Discounted_Price']].head()
```

```python
#Info Tabel Baru
df.info()
```

## Inisialisasi Great Expectations

```python
# Inisialisasi Data Context
context = gx.get_context()

# Membuat Validator dari DataFrame Pandas
validator = context.sources.pandas_default.read_dataframe(df)

print("Validator berhasil dibuat!")
```

## Eksekusi Validasi

### Expectation 1 - Check Unique ID

```python
# Validasi apakah kolom Transaction_ID benar-benar unik.
validation_result = validator.expect_column_values_to_be_unique('Transaction_ID')

print(f"Validation Success: {validation_result['success']}")
print(f"   Jumlah Data Dicek     : {validation_result['result']['element_count']}")
print(f"   Jumlah Duplikat (Fail): {validation_result['result']['unexpected_count']}")
print(f"   Persentase Duplikat   : {validation_result['result']['unexpected_percent']}%")
```

### Expectation 2 - Check Rating Range

```python
# Validasi apakah Rating berada di antara 0 sampai 5.
validation_result = validator.expect_column_values_to_be_between(
    column='Rating', 
    min_value=0, 
    max_value=5
)

print(f"Validation Success    : {validation_result['success']}")
print(f"Jumlah Data Dicek     : {validation_result['result']['element_count']}")
print(f"Jumlah Fail           : {validation_result['result']['unexpected_count']}")
```

### Expectation 3 - Check Category Consistency

```python
# Ambil daftar kategori unik dari data sebagai referensi set
# Validasi apakah Main_Category konsisten dengan daftar kategori yang ada di data saat ini.
valid_cats = list(df['Main_Category'].unique())

validation_result = validator.expect_column_values_to_be_in_set('Main_Category', valid_cats)

print(f"Validation Success : {validation_result['success']}")
print(f"Total Data Dicek   : {validation_result['result']['element_count']}")
print(f"Data Tidak Valid   : {validation_result['result']['unexpected_count']}")
```

### Expectation 4 - Check Price Type (Harus Float)

```python
validation_result = validator.expect_column_values_to_be_in_type_list(
    'Discounted_Price', 
    ['float', 'float64']
)

print(f"Validation Success: {validation_result['success']}")
print(f"Tipe Data Ditemukan: {validation_result['result']['observed_value']}")
```

### Expectation 5 - Check Product_Id Length (Expectation Tambahan)

```python
#Panjang karakter Product_Id harus 10 digit.

validation_result = validator.expect_column_value_lengths_to_equal('Product_Id', 10)

print(f"Validation Success: {validation_result['success']}")
print(f"Jumlah Data Dicek : {validation_result['result']['element_count']}")
print(f"Jumlah Fail       : {validation_result['result']['unexpected_count']}")
```

### Expectation 6 - Check ID Format (Regex) (Expectation Tambahan)

```python
# Format ID hanya boleh Huruf Kapital dan Angka
validation_result = validator.expect_column_values_to_match_regex(
    'Product_Id', 
    r'^[A-Z0-9]+$'
)

print(f"Validation Success: {validation_result['success']}")
print(f"Jumlah Data Dicek : {validation_result['result']['element_count']}")
print(f"Jumlah Fail       : {validation_result['result']['unexpected_count']}")
```

### Expectation 7 - Check Row Count

```python
# Jumlah baris harus di antara 100 dan 10.000. Untuk memastikan jumlah baris wajar (tidak terlalu sedikit atau kebanyakan).
validation_result = validator.expect_table_row_count_to_be_between(
    min_value=100, 
    max_value=10000
)

print(f"Validation Success: {validation_result['success']}")
print(f"Observed Row Count: {validation_result['result']['observed_value']}")
```
